## Data Visualization

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style('whitegrid')

# Read the dataset into a Pandas DataFrame
plot = pd.read_csv('champions_21_22.csv')

# set countplot palette
ax =sns.countplot(x = 'Wins', data = plot, palette='cividis' )

# Chart and axes titles
ax.set_title('Champions League 2021/2022',fontsize = 18, fontweight='bold' )
ax.set_xlabel('Champions League November games', fontsize = 15)
ax.set_ylabel('Teams Wins', fontsize = 15)

 
```

```python

total=float(len(data))
ax = sns.countplot(x="Points", data=data , palette='cividis' ) 

ax.set_xlabel('November games points teams', fontsize = 15)
ax.set_ylabel('Champions League November games', fontsize = 11)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{0:.0%}'.format(height/total),
            ha="center", va='bottom') 

#plt.savefig('snscounter.pdf')
```

```python
# pearson : Standard correlation coefficient
f, ax = plt.subplots(figsize=(20, 10))
corr = data.corr("pearson")
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax,annot=True)
```

Each square shows the correlation between the variables on each axis. Correlation ranges from -1 to +1:

The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is.

A correlation closer to -1 is similar, but instead of both increasing one variable will decrease. 

The diagonals are all 1/dark red because those squares are correlating each variable to itself (so it's a perfect correlation). 

For the rest the larger the number and darker the color the higher the correlation between the two variables. The plot is also symmetrical about the diagonal since the same two variables are being paired together in those squares. 

Values closer to zero means there is no linear trend between the two variables. 

 
From the above corr-plot, it can be inffered in the data set.<br>
The correlation plot shows the relation between the parameters.<br>

1. Wins,Goals for,Goals difference are the most correlated parameters with Points.

2. Loses and Goals againts Function have little correlation with Points.

3. Ties and have tiny correlation with Points.
